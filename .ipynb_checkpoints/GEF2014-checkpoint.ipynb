```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import ensemble
from sklearn.svm import SVR
```

```python
data = pd.read_csv('C:/Users/asuzanovich/Documents/work/gef/GEFCom2014 Data/Load/Task 1/L1-train.csv')
data = data[data['LOAD'].notnull()]
```

```python
data.plot('w1', 'LOAD', 'scatter')
plt.show()
```

```python
data.plot('w1', 'w20', 'scatter')
plt.show()
```

```python
y = data['LOAD']
X = data.drop(['LOAD', 'ZONEID', 'TIMESTAMP'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# use Linear Regression
lr = linear_model.LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
mean_squared_error(y_test, y_pred)
#r2_score(y_test, y_pred)  
```

```python
# plot CV prediction
predicted = cross_val_predict(lr, X, y, cv=10)

fig, ax = plt.subplots()
ax.scatter(y, predicted)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

```python
# GradientBoostingRegressor

params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 2,
          'learning_rate': 0.01, 'loss': 'ls'}
clf = ensemble.GradientBoostingRegressor(**params)
```

```python
clf.fit(X_train, y_train)
mse = mean_squared_error(y_test, clf.predict(X_test))
print("MSE: %.4f" % mse)
```

```python
# plot GradientBoostingRegressor CV prediction

predicted = cross_val_predict(clf, X, y, cv=10)

fig, ax = plt.subplots()
ax.scatter(y, predicted)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```

```python
svr_poly = SVR(kernel='poly', C=1e3, degree=2)
predicted = cross_val_predict(svr_poly, X, y, cv=10)

fig, ax = plt.subplots()
ax.scatter(y, predicted)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Measured')
ax.set_ylabel('Predicted')
plt.show()
```
